```python
MARKSIZE = 8
LINEWIDTH = 3.5
MEAN_LINEWIDTH = 2
PADDING_RATIO = 0.05
ROTATION = 90

import re
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


def parse_eval_log_to_df(log_text: str) -> pd.DataFrame:
    pattern = re.compile(
        r"^\s*(\d+)\s*\|\s*(.*?)\s*\|\s*"
        r"TA=\s*(\d+)\s*TR=\s*(\d+)\(CD=\s*(\d+),DE=\s*(\d+)\)\s*"
        r"FA=\s*(\d+)\s*FR=\s*(\d+)\s*\|\s*"
        r"FAR=\s*([0-9.]+)%\s*FRR=\s*([0-9.]+)%\s*DER=\s*([0-9.]+)%\s*$"
    )

    rows = []
    for line in log_text.splitlines():
        m = pattern.match(line)
        if not m:
            continue
        rows.append(
            {
                "attr": m.group(2).strip(),
                "FAR%": float(m.group(9)),
                "FRR%": float(m.group(10)),
                "DER%": float(m.group(11)),
            }
        )

    if not rows:
        raise ValueError("No valid evaluation lines parsed.")

    return pd.DataFrame(rows)


def plot_far_frr_der_lines_from_log(
    log_text: str,
    sort_by: str = "FAR%",
    ascending: bool = True,
    drop_hyphen_number_attrs: bool = True,
    figsize=(18, 6),
    save_path: str | None = None,   # 🔥 新增
    dpi: int = 300,
):
    plt.rcParams.update({
        "font.size": 13,
        "axes.titlesize": 15,
        "axes.labelsize": 13,
        "legend.fontsize": 12,
    })

    df = parse_eval_log_to_df(log_text)

    # Filter out AAA-1 style attributes
    if drop_hyphen_number_attrs:
        df = df[~df["attr"].str.contains(r"-\d+$", regex=True)].copy()

    df = df.sort_values(sort_by, ascending=ascending).reset_index(drop=True)

    far_mean = df["FAR%"].mean()
    frr_mean = df["FRR%"].mean()
    der_mean = df["DER%"].mean()

    y_min = min(df["FAR%"].min(), df["FRR%"].min(), df["DER%"].min())
    y_max = max(df["FAR%"].max(), df["FRR%"].max(), df["DER%"].max())

    pad = (y_max - y_min) * PADDING_RATIO
    y_lower = max(0, y_min - pad)
    y_upper = y_max + pad

    x = np.arange(len(df))

    fig, ax = plt.subplots(figsize=figsize)

    far_line, = ax.plot(
        x, df["FAR%"],
        linewidth=LINEWIDTH,
        marker="o",
        markersize=MARKSIZE,
        markerfacecolor="white",
        markeredgewidth=1.8,
        label="FAR"
    )

    frr_line, = ax.plot(
        x, df["FRR%"],
        linewidth=LINEWIDTH,
        marker="s",
        markersize=MARKSIZE,
        markerfacecolor="white",
        markeredgewidth=1.8,
        label="FRR"
    )

    der_line, = ax.plot(
        x, df["DER%"],
        linewidth=LINEWIDTH,
        marker="^",
        markersize=MARKSIZE,
        markerfacecolor="white",
        markeredgewidth=1.8,
        label="DER"
    )

    ax.axhline(far_mean, linestyle="--", linewidth=MEAN_LINEWIDTH, color=far_line.get_color())
    ax.axhline(frr_mean, linestyle="--", linewidth=MEAN_LINEWIDTH, color=frr_line.get_color())
    ax.axhline(der_mean, linestyle="--", linewidth=MEAN_LINEWIDTH, color=der_line.get_color())

    ax.set_xticks(x)
    ax.set_xticklabels(df["attr"], rotation=ROTATION)
    ax.set_ylabel("Error Rate (%)")
    ax.set_xlabel(f"Attribute (Sorted by {sort_by}, low→high)")
    ax.set_title("Attribute-Level FAR / FRR / DER")

    ax.set_ylim(y_lower, y_upper)

    ax.grid(True, which="major", linestyle="--", alpha=0.25)
    ax.minorticks_on()
    ax.grid(True, which="minor", linestyle=":", alpha=0.15)

    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)

    ax.legend(frameon=False, ncol=3)

    fig.tight_layout()

    # 🔥 保存
    if save_path is not None:
        fig.savefig(save_path, dpi=dpi, bbox_inches="tight")
        print(f"Figure saved to {save_path}")

    plt.show()

    return df
```
